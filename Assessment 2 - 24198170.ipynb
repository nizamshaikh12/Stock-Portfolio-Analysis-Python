#Part 2: Control Flow for Trading Decisions

#Exercise 2.1: if/elif/else Statements

```python
#1
stock = 'NVDA'
#2
price = stock_details[stock]
if price > 100: 
    print('sell this stock, its overvalued!')
elif 100 <= price <= 200:
    print('hold this stock, its stable.')
else:
    print('buy more of this stock, its bargain!')
```

#Exercise 2.2: for Loop

```python
#1
stock_details = {'AAPL': 254.43, 'GOOG': 251.66, 'MSFT': 509.23, 'NVDA': 178.43, 'TSLA': 425.85, 'INTC': 24.82, 'META': 743.75}
for ticker,price in stock_details.items():
    print(f'The price of {ticker} is :${price} ')
```

#Exercise 2.3: while Loop

```python
#1
target_price = 250
```

```python
#2
current_price = stock_details['NVDA']
current_price
```

```python
#3, #4
while current_price < target_price:
    print(f'current_price: ${current_price}')
    current_price += 5
```

```python
#5
print(f'The stock has reached its target price of ${target_price}')
```
