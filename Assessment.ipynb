```python
# defining vowels in name 
def Count_Vowels(name):
    total = 0
    vowels = ['a','e','i','o','u']
    for i in name.lower():
        if i in vowels:
            total += 1
    return total
```

```python
#defining percent of vowels
def Percent_Vowels(name) :
    num_vowels = Count_Vowels(name)
    percent = num_vowels/len(name)
    return percent
```
