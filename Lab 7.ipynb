```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Stock Tickers
# --- Portfolio Data ---
# Stock Tickers
tickers = ['MSFT', 'GOOG', 'TSLA', 'AMZN', 'NVDA',
'NFLX', 'DIS', 'PYPL', 'ADBE', 'CRM']
# Annual Return Percentage (e.g., 0.15 is 15%)
annual_returns = np.array([0.15, 0.08, 0.35, -0.05, 0.50,
0.12, 0.01, -0.10, 0.22, 0.18])
# Stock Volatility (Standard Deviation)
volatility = np.array([0.20, 0.15, 0.40, 0.25, 0.55,
0.28, 0.18, 0.30, 0.24, 0.21])
# Investment Allocation (Total must equal 1 or 100%)
allocation_weights = np.array([0.15, 0.10, 0.05, 0.10, 0.05,
0.15, 0.10, 0.05, 0.15, 0.10])
```

```python
plt.figure(figsize=(10,6))
colors=['g' if r>0 else 'r' for r in annual_returns]
plt.bar(tickers,annual_returns*100, color=colors)
plt.xlabel("Stock Tickers")
plt.ylabel("Annnual Return (%)")
plt.title("Portfolio Stock Annual Returns")
plt.axhline(0, color='gray', linewidth=0.8) # add a zero line
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10,6))
scatter=plt.scatter(volatility,annual_returns,s=allocation_weights*5000,#set seze based on investment weight 
                    c=annual_returns,#set color based onreturn value
                    cmap='RdYlGn',#Use a red-yellow-green color map
                    alpha=0.7,edgecolors='w', linewidth=1)
plt.xlabel("Volatility (Risk-Standard Deviation)")
plt.ylabel("Annual return (Reward)")
plt.title("Risk vs.Reward in Portfolio")

#Find and annotate the highest return stock(NVDA)
max_return_index=np.argmax(annual_returns)
plt.annotate(tickers[max_return_index],
             (volatility[max_return_index],
annual_returns[max_return_index]),
             textcoords="offset points",xytext=(5,-10),ha='center',
             arrowprops=dict(arrowstyle="->",
connectionstyle="arc3,rad=.2"))
plt.colorbar(scatter,label='Annual Return Value')
plt.grid(True,linestyle=':',alpha=0.5)
plt.show()
              
              
```

```python
plt.figure(figsize=(8, 5))
plt.hist(annual_returns, bins=5, edgecolor='black', alpha=0.9,
color='skyblue')
plt.xlabel("Annual Return Value")
plt.ylabel("Number of Stocks (Frequency)")
plt.title("Distribution of Portfolio Returns")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
# Create an array to explode the largest slice (MSFT is at index 0, 0.15)
explode_arr = [0.1 if i == 0 else 0 for i in range(len(tickers))]
plt.pie(allocation_weights,
labels=tickers,
autopct='%1.1f%%',
startangle=90,
explode=explode_arr,
shadow=True)
plt.title("Portfolio Investment Allocation by Ticker")
plt.axis('equal') # Ensures the pie chart is circular
plt.show()
```
